```python
from datetime import datetime, timedelta

from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, Range1d, DatetimeTickFormatter

from mrmat_python_vis_bokeh.model import Timeseries

t = Timeseries(datafile='../../build/test.json')

output_notebook()

now = datetime.utcnow()
last_day = now - timedelta(days=1)
last_week = now - timedelta(weeks=1)
last_month = now - timedelta(weeks=4)

ds_daily = ColumnDataSource(t.daily())
f_daily = figure(title=f'Last day: {last_day.strftime("%d %b %y %H:%M")} - {now.strftime("%d %b %y %H:%M")}',
                 plot_width=800, plot_height=300,
                 x_axis_label='Time (UTC)',
                 y_axis_label='Seconds',
                 x_range=Range1d(last_day, now),
                 y_range=Range1d(0, 11),
                 x_axis_type='datetime')
f_daily.xaxis[0].formatter = DatetimeTickFormatter(hours=['%H:%M'], days='%d %b %y')

f_daily.line(source=ds_daily, x='index', y='foo', color='blue')
f_daily.circle(source=ds_daily, x='index', y='foo', line_color='black', fill_color='white', size=3)
show(f_daily)

ds_weekly = ColumnDataSource(t.weekly())
f_weekly = figure(title=f'Last Week: {last_week.strftime("%d %b %y")} - {now.strftime("%d %b %y")}',
                  plot_width=800, plot_height=300,
                  x_axis_label='Time (UTC)',
                  y_axis_label='Seconds',
                  x_range=Range1d(last_week, now),
                  y_range=Range1d(0,11),
                  x_axis_type='datetime')
f_weekly.xaxis[0].formatter = DatetimeTickFormatter(hours=['%H:%M'], days=['%a'])
f_weekly.line(source=ds_weekly, x='index', y='foo', color='blue')
f_weekly.circle(source=ds_weekly, x='index', y='foo', line_color='black', fill_color='white', size=3)
show(f_weekly)

ds_monthly = ColumnDataSource(t.monthly())
f_monthly = figure(title=f'Last Month: {last_month.strftime("%d %b %y")} - {now.strftime("%d %b %y")}',
                  plot_width=800, plot_height=300,
                  x_axis_label='Time (UTC)',
                  y_axis_label='Seconds',
                  x_range=Range1d(last_month, now),
                  y_range=Range1d(0,11),
                  x_axis_type='datetime')
f_monthly.xaxis[0].formatter = DatetimeTickFormatter(days=['%d %b %y'])
f_monthly.line(source=ds_monthly, x='index', y='foo', color='blue')
f_monthly.circle(source=ds_monthly, x='index', y='foo', line_color='black', fill_color='white', size=3)

show(f_monthly)
```
